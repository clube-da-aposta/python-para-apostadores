**Exerccio 1:** Pea o usurio para digitar dois nmeros, crie uma funo que recebe esses dois nmeros e utilize funes annimas para realizar as seguintes operaes: soma, subtrao, multiplicao, diviso real. Essa funo no deve retornar nada e mostrar o resultados das operaes.

````python
def exibe(a, b):
    soma = lambda a, b: a + b
    subt = lambda a, b: a - b
    mult = lambda a, b: a * b
    divi = lambda a, b: a / b
    print(f'A soma dos nmeros digitados : {round(soma(a, b),2)}')
    print(f'A diferena dos nmeros digitados : {round(subt(a, b),2)}')
    print(f'O produto dos nmeros digitados : {round(mult(a, b),2)}')
    print(f'A diviso dos nmeros digitados : {round(divi(a,b),2)}')

num1 = float(input('Digite o primeiro nmero: '))
num2 = float(input('Digite o segundo nmero: '))

exibe(num1, num2)
````

**Exerccio 2:** Thiago armazenou os valores apostados em um determinado ms na lista abaixo. Crie uma funo que calcule a mdia que ele apostou nesse ms. A funo deve retornar o valor da mdia, exiba esse valor na tela para o usurio<br> Use depois a funo filter para mostrar em ordem crescente quais valores so iguais ou esto abaixo da mdia e quais esto acima.<br>
```
apostas = [57.31, 114.57, 96.39, 112.19, 72.25, 157.87, 187.94, 57.64, 103.94, 164.91, 93.51, 137.93, 68.72, 122.63]
```

````python
def media(apostas):
    return round(sum(apostas)/len(apostas),2)

apostas = [57.31, 114.57, 96.39, 112.19, 72.25, 157.87, 187.94, 57.64, 103.94, 164.91, 93.51, 137.93, 68.72, 122.63]
m = media(apostas)
abaixo = list(filter(lambda x: x <= m, apostas))
acima = list(filter(lambda x: x > m, apostas))
print(f'Thiago apostou em mdia nesse ms o valor de: R${m}')
print(f'Valores abaixo ou iguais da mdia: {sorted(abaixo)}')
print(f'Valores acima da mdia: {sorted(acima)}')
````

**Exerccio 3:** Thiago colocou os valores de algumas apostas em dicionrios distintos. Nos quais, o ms da aposta se encontra no nome do dicionrio, as chaves do dicionrio so os dias do ms no qual as apostas foram realizadas, e os valores so listas contendo o valor apostado, a odd e se ele ganhou ou perdeu. Crie um funo que ajude Thiago a calcular os ganhos que ele obteve em cada ms, a funo deve retornar um dicionrio com os ganhos daquel ms especfico. Faa uma nova funo receba uma lista com vrios dicionrios e junte-os em um nico dicionrio. Crie uma terceira funo que no retorne nada, mas exiba uma mensagem na tela informando o ms e os ganhos ou perdas obtidos.<br>

```
jan = {
'ms': 'Janeiro',
5: [86.72, 2.37, 'Ganha'], 
14: [79.9, 1.81, 'Ganha'], 
25: [182.61, 1.67, 'Perdida']} 

fev = {
'ms': 'Fevereiro',
3: [98.5, 1.84, 'Ganha'],

27: [126.97, 2.51, 'Ganha']} 

mar = {
'ms': 'Maro',
1: [124.31, 2.7, 'Perdida'],

'ms': 'Abril',
2: [145.69, 1.64, 'Ganha'],

mai = {
'ms': 'Maio',

````python
def ganho(aposta_dict):
    '''
    Funo que recebe um dicionrio e calcula os ganhos para um determinado ms
    O dicionrio deve estar com a seguinte formatao:
    jan = {'ms': 'Janeiro', 5: [86.72, 2.37, 'Ganha'], 14: [79.9, 1.81, 'Ganha'], 25: [182.61, 1.67, 'Perdida']}
    No qual, so includos o nome do ms, e as demais chaves representam dias daquele ms.
    Cada uma possui uma lista com os valores da aposta, odd, se ganhou ou perdeu, nessa ordem
    '''
    lista_ganhos = []
    ganho_dict = {}
    mes = aposta_dict.pop('ms')
    for value in aposta_dict.values():
        if value[2] == 'Ganha':
            r = value[0] * (value[1] - 1)
            lista_ganhos.append(r)
        else:
            r = -value[0]
            lista_ganhos.append(r)
    ganho_dict[mes] = round(sum(lista_ganhos),2)
    lista_ganhos.clear()
    return ganho_dict

def junta(lista_dict):
    final_dict = {}
    for dict_ in lista_dict:
        final_dict.update(dict_)
    return final_dict

def exibe(dic_):
    for key, value in dic_.items():
        print(f'Thiago ganhou/perdeu R${value} no ms de {key}.')
    return None

jan = {'ms': 'Janeiro', 5: [86.72, 2.37, 'Ganha'], 14: [79.9, 1.81, 'Ganha'], 25: [182.61, 1.67, 'Perdida']} 
fev = {'ms': 'Fevereiro', 3: [98.5, 1.84, 'Ganha'], 11: [190.46, 2.02, 'Perdida'], 27: [126.97, 2.51, 'Ganha']} 
mar = {'ms': 'Maro', 1: [124.31, 2.7, 'Perdida'], 15: [156.24, 1.07, 'Ganha'], 29: [68.81, 2.07, 'Ganha']}
abr = {'ms': 'Abril', 2: [145.69, 1.64, 'Ganha'], 12: [115.29, 1.79, 'Perdida'], 30: [140.65, 1.39, 'Ganha']} 
mai = {'ms': 'Maio', 3: [139.9, 2.81, 'Ganha'], 10: [192.68, 2.77, 'Perdida'], 15: [76.24, 1.05, 'Perdida'], 31: [110.78, 2.31, 'Perdida']}

lista_dict = [ganho(jan), ganho(fev), ganho(mar), ganho(abr), ganho(mai)]
final_dict = junta(lista_dict)
exibe(final_dict)
````
